```python
import pandas as pd 
```

```python
data=pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
print(data.shape)
data["Geography"].unique()
```

```python
data_X=data.iloc[:,3:13]
data_X.head()
```

```python
data_X.sample(5)
```

```python
Y=data.iloc[:,-1]
Y.head()
```

```python
geography=pd.get_dummies(data_X.Geography,drop_first=True)
gender=pd.get_dummies(data_X.Gender,drop_first=True)
```

```python
geography.sample(5)
```

```python
gender.head()
```

```python
X=pd.concat([data_X,geography,gender],axis=1)
```

```python
X.head()
```

```python
X=X.drop(['Geography','Gender'],axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,random_state=7,train_size=0.8)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
SD_X_train=sc.fit_transform(X_train)
SD_X_test=sc.transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout
```

```python
classifier=Sequential()
```

```python
#input_dim=no. of input nuerons (no. of incoming features)

classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu',input_dim=11))
classifier.add(Dropout(0.3))
```

```python
classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu'))
classifier.add(Dropout(0.4))
```

```python
classifier.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
classifier.add(Dropout(0.2))
```

```python
classifier.summary()
```

```python
classifier.compile(optimizer='Adamax',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model_his=classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=100)
```

```python
y_hat=classifier.predict(X_test)
y_hat=y_hat>0.5
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_hat,y_test)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_hat)
score
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_test,y_hat)
```
